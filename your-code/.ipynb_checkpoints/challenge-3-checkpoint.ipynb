#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
bins = [10, 20, 50]

for i in range(len(bins)):
    plt.subplot(1,3,i+1)
    titanic['Age'].hist(bins = bins[i])
```

#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
gender = titanic.groupby('Gender').agg({'Gender':'count'}).rename(columns={'Gender':'Gender_count'}).reset_index()
plt.bar(gender['Gender'], gender['Gender_count'], alpha = 0.45, color=['pink','blue'])
plt.title('Titanic gender count', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number', fontsize=12)
plt.show()
```

```python
# your code here- 2nd way
x = gender['Gender']
y = gender['Gender_count']
sns.barplot(x = x, y = y)
```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
# your code here
pclass = titanic.groupby('Pclass').agg({'Pclass':'count'}).rename(columns={'Pclass':'Pclass_count'}).reset_index()
plt.bar(pclass['Pclass'], pclass['Pclass_count'], alpha = 0.4, color = ['purple','black','orange'])

plt.xlabel('Class')
plt.xticks(list(range(4)), ['', '1','2','3'])
plt.ylabel('Count')
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way
data = titanic[['Age','Fare']]
cmap = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(data.corr(),cmap=cmap,annot =True)
```

```python
# your code-2nd way
x = titanic['Fare']
y = titanic['Age']

plt.scatter(x,y)
```
